```r
library(ismev)
library(extRemes)
library(dplyr)

csv <- read.csv("../../data/data_preparation/concat.csv")
csvpath <- "../../data/analysis"
if (!dir.exists(csvpath)) {
    dir.create(csvpath)
}
imagepath <- "../../data/analysis/img"
if (!dir.exists(imagepath)) {
    dir.create(imagepath)
}

df <- data.frame()
diffs <- c("hyper", "another", "leggendaria")
for (diff in diffs) {
    col <- paste(diff, "Level", sep="")
    scores <- csv[!is.na(csv[col]) & csv[col] == 12,c("title", paste(diff, "Notes", sep=""), paste(toupper(diff), ".EXスコア", sep=""), paste(toupper(diff), ".順位", sep=""))]
    names(scores) <- c("title", "notes", "exscore", "rank")
    scores <- scores[scores["rank"] <= 10,]
    fit <- tapply(scores$exscore, scores$title, function(x) { return (fevd(x, type="GEV")) })

    for (i in 1:length(fit)) {
        name <- rownames(fit)[i]
        png(paste(imagepath, "/", gsub("/", "", name), ".png", sep=""))
        plot.fevd(fit[[i]], type = "density", main = paste(name, "[", diff, "]", sep=""), xlab="EX SCORE")
        dev.off()
        p <- fit[[i]]$results$par
        ultimetScoreRate <- qevd(0.99, loc = p[1], scale = p[2], shape = p[3])
        max <- max(scores[scores["title"] == name,]$notes)*2
        if (ultimetScoreRate > max) {
            ultimetScoreRate <- max
        }
        df <- rbind(df, c(paste(name, "[", diff, "]", sep=""), max(scores[scores["title"] == name,]$exscore), ultimetScoreRate, max))
    }
}
names(df) <- c("title", "top", "ult", "max")

write.csv(df, file = paste(csvpath, "/ult.csv", sep=""), row.names=FALSE)
```
